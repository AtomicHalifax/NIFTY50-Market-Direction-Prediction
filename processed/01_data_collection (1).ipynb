```python
def clean_yfinance_columns(df):
    """
    Convert yfinance MultiIndex columns into standard single-level columns.
    """

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    return df
```

```python
def download_market_data(ticker, asset_name):
    """
    Downloads historical market data from Yahoo Finance,
    standardizes the columns, saves the raw CSV,
    and returns a cleaned DataFrame.

    Parameters
    ----------
    ticker : str
        Yahoo Finance ticker symbol.

    asset_name : str
        Name used when saving the dataset.

    Returns
    -------
    pandas.DataFrame
        Cleaned historical market data.
    """

    df = yf.download(
    ticker,
    start="2007-01-01",
    progress=False,
    auto_adjust=False
)


    df = clean_yfinance_columns(df)

    df.columns.name = None

    if df.empty:
        raise ValueError(f"No data downloaded for ticker: {ticker}")

    file_path = f"data/raw/{asset_name}_raw.csv"

    df.to_csv(file_path, index=True)

    print(f"Saved dataset to: {file_path}")

    return df
```
